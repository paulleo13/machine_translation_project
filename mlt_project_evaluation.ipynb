````python
# calculate bleu of sentences
def calculate_bleu(inputs, references, return_all=False):
  """Calculates bleu score for input sentences.

  Args:
    inputs: One or multiple strings whos bleu scores should be calculated.
    references: One or multiple strings as references for calculation.
    n_grams: Value (int) from 1 to 4 determining the number of n-grams 
      on which base bleu will be calculated.
    return_all: Whether to return bleu scores for all sentences or just average.
      Default is only average score.

  Returns:
    The bleu score as float value for provided sentences.
  
  Raises:
    AssertionError if inputs and references are of different length and 
      ngrams are not between 1 and 4
  
  """
  sentence_pairs = zip(inputs, references)
  bleu_scores = [sentence_bleu(nltk.word_tokenize(pair[0]), nltk.word_tokenize(pair[1])) for pair in sentence_pairs]
  avg_bleu_score = sum(bleu_scores) / len(bleu_scores)
  if return_all:
    return (avg_bleu_score, bleu_scores)
  return avg_bleu_score
````

````python
from operator import xor
from re import X
# longest common subsequence #dauert zu lange
def rouge_l_calculation(*inputs, references, return_all=False):
  """Calculates the longest common subsequence (lcs) of two sentences.
  
  Args:
    inputs: One or more strings for the lcs calculation.
    references: One or more strings as reference for lcs calculation.
    return_all: Whether to return lcs values for all sentences plus average.

  Returns:
    The average lcs value over all sentences. If return_all True,
    then tuple of average and list of all lcs values.
  
  Raises:
    AssertionError if the lengths of inputs do not match.

  """

  
  assert len(inputs) == len(references)
  sentence_pairs = zip(inputs, references)  
  

  lcs_values = [lcs_2(pair[0].split(), pair[1].split()) for pair in sentence_pairs]
  avg_lcs_val = sum(lcs_values) / len(lcs_values)
  if return_all:
    return (avg_lcs_val, lcs_values)
  return avg_lcs_val


def lcs_2(pattern_1, pattern_2):
    m = len(pattern_1)
    n = len(pattern_2)
    # dp will store solutions as the iteration goes on
    dp = [[None] * (n + 1) for item in range(m + 1)]
    for i in range(m + 1):
        for j in range(n + 1):
          if i == 0 or j == 0:
            dp[i][j] = 0
          elif pattern_1[i - 1] == pattern_2[j - 1]:
              dp[i][j] = dp[i - 1][j - 1] + 1
          else :
              dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]
````

````python
from statistics import mean
def average_length_of_sentence(sentences):
  lengths = []
  for sentence in sentences:
    lengths.append(len(sentence.split(" ")))   
  return mean(lengths)
````

````python
# single word count #
def single_word_count(*sentences, filter_stop_words=True, language):
  """Counts all single words. 

  Args:
    sentences: The sentences of which to count the words.
    filter_stop_words: Filter all words with no contextual information (e.g. "and", "a", "the", [...]). Default is True.
    language: Which language the sentences are in. Either "english" or "spanish".

  Returns:
    Dictionary of words and their respective occurrences (e.g. {"house": 54, "dog": 23}).
  
  """
  stoplist = stopwords.words(language)
  words_dict = {}
  for sentence in sentences:
    words = sentence.split()
    for word in words:
      if word in words_dict.keys():
        words_dict[word] += 1
      else:
        words_dict.update({word, 1})
  return words_dict
````

````python
# word and char bigram and trigram count
## DO NOT CALL WITH ALL SENTENCES -> THE RAM WILL FILL UP AND THE KERNEL DIE! USE BATCH FUNCTION AT BOTTOM ##

def word_gram_count(sentences, language, count_bigrams=True, count_trigrams=True, filter_stop_words=True):
  """Counts word bi- and/ or trigrams in sentences.

  Args:
    sentences: One or more strings whos n-grams are to be counted.
    count_bigrams: If true, bigrams are counted.
    count_trigrams: If true, trigrams are counted.
    filter_stop_words: Filter all words with no contextual information (e.g. "and", "a", "the", [...]). Default is True.
    language: Which language the sentences are in. Either "english" or "spanish".

  Returns:
    Pandas DataFrame containing the information.
  
  Raises:
    AssertionError if the language does not match one of the available options.
    ValueError if both count_bigrams and count_trigrams are False.

  """
  assert language == "english" or language == "spanish"
  if not count_bigrams and not count_trigrams:
    raise ValueError("Function has to count either bi- and/ or trigrams. They cannot both be False.")
  stoplist = stopwords.words(language)
  if not count_bigrams and count_trigrams:
    c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(3,3))
  elif count_bigrams and not count_trigrams:
    c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(2,2))
  else:
    c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(2,3))
  # create dataframe
  df = to_pd_dataframe(sentences)
  # matrix of ngrams
  ngrams = c_vec.fit_transform(df['Sentences'])
  # count frequency of ngrams
  count_values = ngrams.toarray().sum(axis=0)
  # list of ngrams
  vocab = c_vec.vocabulary_
  df_ngram = pd.DataFrame(sorted([(count_values[i],k) for k,i in vocab.items()], reverse=True)
            ).rename(columns={0: 'frequency', 1:'bigram/trigram'})
  return df_ngram


def char_gram_count(sentences, language, count_bigrams=True, count_trigrams=True, filter_stop_words=True):
  """Counts char bi- and/ or trigrams in sentences.

  Args:
    sentences: One or more strings whos n-grams are to be counted.
    count_bigrams: If true, bigrams are counted.
    count_trigrams: If true, trigrams are counted.
    filter_stop_words: Filter all words with no contextual information (e.g. "and", "a", "the", [...]). Default is True.
    language: Which language the sentences are in. Either "english" or "spanish".

  Returns:
    Pandas DataFrame containing the information.
  
  Raises:
    AssertionError if the language does not match one of the available options.
    ValueError if both count_bigrams and count_trigrams are False.

  """
  assert language == "english" or language == "spanish"
  if not count_bigrams and not count_trigrams:
    raise ValueError("Function has to count either bi- and/ or trigrams. They cannot both be False.")
  stoplist = stopwords.words(language)
  if not count_bigrams and count_trigrams:
    c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(3,3), analyzer='char')
  elif count_bigrams and not count_trigrams:
    c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(2,2), analyzer='char')
  else:
    c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(2,3), analyzer='char')
  # create dataframe
  df = to_pd_dataframe(sentences)
  # matrix of ngrams
  ngrams = c_vec.fit_transform(df['Sentences'])
  # count frequency of ngrams
  count_values = ngrams.toarray().sum(axis=0)
  # list of ngrams
  vocab = c_vec.vocabulary_
  df_ngram = pd.DataFrame(sorted([(count_values[i],k) for k,i in vocab.items()], reverse=True)
            ).rename(columns={0: 'frequency', 1:'bigram/trigram'})
  return df_ngram
````

````python

def lexical_diversity(sentences, language):
  long_sentence = ""
  for sentence in sentences:
    long_sentence = long_sentence + " " + sentence

  stop_words = set(stopwords.words(language))

  tokenizer = RegexpTokenizer(r'\w+')
  word_tokens = tokenizer.tokenize(long_sentence)

  filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words]

  """filtered_sentence = []

  for w in word_tokens:
    if w not in stop_words:
      filtered_sentence.append(w)
"""
  word_count = len(filtered_sentence)
  vocab_size = len(set(filtered_sentence))
  diversity_score = vocab_size / word_count
  return diversity_score
````

````python
# divide data into batches and execute a function for all batches
def create_batches(data, batchsize):
  """Divides a batch of data into batches of given size.

  Args:
    data: The data to split in list format.
    batchsize: Size of desired batches.

  Returns:
    Iterable of all batches with batchsize.
  
  Raises:
    AssertionError if the size of the data is not a multiple of batchsize.

  """
  assert len(data) % batchsize == 0
  batches = []
  num_batches = len(data) // batchsize
  for i in range(num_batches):
    batches.append(data[i*batchsize:(i+1)*batchsize])
  return batches
````

````python
from collections import Counter
import nltk

def count_tags(sentences):
  long_sentence = ""
  for sentence in sentences:
    long_sentence = long_sentence + " " + sentence

  lower_case = long_sentence.lower()
  tokens = nltk.word_tokenize(lower_case)
  tags = nltk.pos_tag(tokens)
  counts = Counter( tag for word,  tag in tags)
  print(counts)
````
